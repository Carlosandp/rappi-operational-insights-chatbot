# Rappi Ops Analytics — Exploratory Data Analysis

Run this notebook to explore the raw data before running the Streamlit app.

```python
import sys
sys.path.insert(0, '../app')

import pandas as pd
import numpy as np
import plotly.express as px

from utils import get_processed_data

data = get_processed_data()
mw = data['metrics_wide']
ow = data['orders_wide']
ml = data['metrics_long']
ol = data['orders_long']

print('Metrics wide:', mw.shape)
print('Orders wide:', ow.shape)
mw.head(3)
```

```python
# Zone distribution by country
zone_counts = mw.groupby('COUNTRY')['ZONE'].nunique().sort_values(ascending=False)
fig = px.bar(zone_counts.reset_index(), x='COUNTRY', y='ZONE', title='Zones per Country')
fig.show()
```

```python
# Perfect Orders distribution
po = mw[mw['METRIC'] == 'Perfect Orders']
fig = px.histogram(po, x='L0W', color='ZONE_TYPE', nbins=50,
                   title='Perfect Orders Distribution (L0W) by Zone Type')
fig.show()
```

```python
# Correlation heatmap
pivot = mw.pivot_table(index=['COUNTRY','CITY','ZONE'], columns='METRIC', values='L0W').dropna(how='all')
corr = pivot.corr()
fig = px.imshow(corr, title='Metric Correlations (L0W)', color_continuous_scale='RdBu_r',
                zmin=-1, zmax=1)
fig.show()
```
